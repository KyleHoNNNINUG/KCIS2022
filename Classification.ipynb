```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_data = load_breast_cancer()
print(my_data)

# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_breast_cancer.html
```

```python
print(my_data.keys())
```

```python
print("資料筆數：", my_data)
```

```python
df = pd.DataFrame(my_data.data, columns = my_data.feature_names)

df['target'] = my_data['target']
print(df)
```

```python
train_x, test_x, train_y, test_y = train_test_split(my_data.data, my_data.target, test_size=0.2, random_state=1001, shuffle=True)

print("原始資料集的維度大小：", my_data.data.shape)
print("訓練集的維度大小：", train_x.shape)
print("測試資料集的維度大小：", test_x.shape)
```

```python
my_model = KNeighborsClassifier()

my_model.fit(train_x, train_y)

test_score = my_model.score(test_x, test_y)
print("模型評估完測試集的準確度為：", test_score)
```

```python
plt.scatter(train_x[:,0], train_x[:,1], alpha=0.2)
```

```python
plt.scatter(train_x[:,0], train_x[:,1], alpha=0.5, c=my_model.predict(train_x))
```

```python
my_model = KNeighborsClassifier(n_neighbors=9)

my_model.fit(train_x, train_y)

test_score = my_model.score(test_x, test_y)
print("模型評估完測試集的準確度為：", test_score)

plt.scatter(train_x[:,0], train_x[:,1], alpha=0.5, c=my_model.predict(train_x))
```
